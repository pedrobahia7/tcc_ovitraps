```python
import pandas as pd 
final_df = pd.read_csv('../data/final_data.csv')
```

```python
final_df['dtcol'] = pd.to_datetime(final_df['dtcol'])
final_df['dtinstal'] = pd.to_datetime(final_df['dtinstal'])
```

```python
print('Max dtinstal',final_df['dtinstal'].max())
print('Min dtinstal',final_df['dtinstal'].min())
print('Max dtcol',final_df['dtcol'].max())
print('Min dtcol',final_df['dtcol'].min())
```

Note that the problem is on dtcol, since the distance between dtcol and the next dtinstal is always bigger than 10 

```python
for _,row in final_df[final_df['dtcol'] == final_df['dtinstal']][['narmad','dtcol','dtinstal']].iterrows():
    next_instal = final_df.loc[(final_df['narmad'] == row['narmad']) & (final_df['dtcol'] > row['dtcol']),['dtinstal','dtcol']].sort_values('dtcol').iloc[0:1]['dtinstal']
    previous_col = final_df.loc[(final_df['narmad'] == row['narmad']) & (final_df['dtinstal'] < row['dtinstal']),['dtinstal','dtcol']].sort_values('dtcol').iloc[-1:]['dtcol']
    print('Narmad:',row['narmad'])
    print('dtcol', next_instal - row['dtcol'] )
    print('\n','dtinstal', row['dtinstal'] - previous_col )
    print('\n')
```

```python
# Correcting the dates
final_df.loc[final_df['dtcol'] == final_df['dtinstal'],'dtcol'] = final_df.loc[final_df['dtcol'] == final_df['dtinstal'],'dtcol'] + pd.Timedelta(days=7)
```

```python
# save final df
final_df.to_csv('../data/final_data.csv',index=False)
```
