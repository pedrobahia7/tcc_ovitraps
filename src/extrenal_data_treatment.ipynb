INMET_auto (Pampulha, Rola Moca e  Cercadinho)


```python
root = '../data/external_data/INMET_auto/Rola_Moca/'
final_df = pd.DataFrame()

for file in os.listdir(root):     
    raw_data = pd.read_csv(root + file,
                        sep=';',skiprows=8, encoding='latin1')
    try:
        raw_data.set_index(pd.to_datetime(raw_data["Data"] + " " + raw_data["Hora UTC"], format="%Y/%m/%d %H%M UTC"),inplace = True) 
        final_data = raw_data[['PRECIPITAÃÃO TOTAL, HORÃRIO (mm)','TEMPERATURA DO AR - BULBO SECO, HORARIA (Â°C)','UMIDADE RELATIVA DO AR, HORARIA (%)']]
   
    except:
        raw_data.set_index(pd.to_datetime(raw_data["DATA (YYYY-MM-DD)"] + " " + raw_data["HORA (UTC)"], format="%Y-%m-%d %H:%M"),inplace = True)
        final_data = raw_data[['PRECIPITAÃÃO TOTAL, HORÃRIO (mm)','TEMPERATURA DO AR - BULBO SECO, HORARIA (Â°C)','UMIDADE RELATIVA DO AR, HORARIA (%)']]
    final_data.columns = ['Precipitacao','Temperatura','Umidade']
    final_data['latitude'] = -20.0311111
    final_data['longitude'] = -44.01111111
    final_data = final_data.applymap(lambda x: x.replace(',','.') if type(x) == str else x)
    final_data = final_data.astype(float)
    final_df = pd.concat([final_data,final_df],axis=0)
final_df.sort_index(inplace=True)
final_df.to_csv('../data/external_data/estacao_rola_moca.csv')   
```

INMET_map (Aeroporto Pampulha)

```python
root = '../data/external_data/INMET_map/'
final_df = pd.DataFrame()

for file in os.listdir(root):     
    raw_data = pd.read_csv(root + file,
                        sep=';')
    raw_data.set_index(pd.to_datetime(raw_data['Data'] + ' ' + raw_data['Hora (UTC)'].apply(lambda x: f"{x:04d}").str[:2] + ':' + raw_data['Hora (UTC)'].apply(lambda x: f"{x:04d}").str[2:]),inplace = True)
    final_data = raw_data[['Chuva [Diaria] (mm)','Temp. [Hora] (C)','Umi. (%)']]
    final_data.columns = ['Precipitacao','Temperatura','Umidade']
    final_data['latitude'] = -19.85
    final_data['longitude'] = -43.95
    final_data = final_data.applymap(lambda x: x.replace(',','.') if type(x) == str else x)
    final_data = final_data.astype(float)
    final_df = pd.concat([final_data,final_df],axis=0)

final_df.sort_index(inplace=True)
final_df.to_csv('../data/external_data/estacao_aeroporto_pampulha.csv')   
```

DECEA-API (Aeroporto Carlos Prates)   

```python

# Read the JSON file first
with open('../data/external_data/DECEA_API/umidade.json', 'r') as file:
    json_data = json.load(file)


umidade_df = pd.DataFrame(json_data['data'][1]['data'])
umidade_df.columns = ['Data','Umidade']
umidade_df['Data'] =umidade_df['Data'].apply(lambda x: datetime.utcfromtimestamp(x/1000).strftime('%Y-%m-%d %H:%M:%S'))
umidade_df.set_index('Data',inplace=True)
umidade_df.index.name = ''  


# Read the JSON file first
with open('../data/external_data/DECEA_API/temperatura.json', 'r') as file:
    json_data = json.load(file)


temperatura_df = pd.DataFrame(json_data['data'][1]['data'])
temperatura_df.columns = ['Data','Temperatura']
temperatura_df['Data'] =temperatura_df['Data'].apply(lambda x: datetime.utcfromtimestamp(x/1000).strftime('%Y-%m-%d %H:%M:%S'))
temperatura_df.set_index('Data',inplace=True)
temperatura_df.index.name = ''  



# Precipitacao
with open('../data/external_data/DECEA_API/precipitacao.json', 'r') as file:
    json_data = json.load(file)

precipitacao_df = pd.DataFrame(json_data['data'][1]['data'])
precipitacao_df.columns = ['Data','Precipitacao']
precipitacao_df['Data'] =precipitacao_df['Data'].apply(lambda x: datetime.utcfromtimestamp(x/1000).strftime('%Y-%m-%d %H:%M:%S'))
precipitacao_df.set_index('Data',inplace=True)
precipitacao_df.index.name = ''  


final_df = pd.concat([precipitacao_df,temperatura_df,umidade_df],axis=1)
final_df['latitude'] = -19.90
final_df['longitude'] = -43.98

final_df.to_csv('../data/external_data/estacao_aeroporto_carlos_prates.csv')   

```
