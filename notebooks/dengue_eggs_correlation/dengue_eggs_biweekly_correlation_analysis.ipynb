```python
import pandas as pd
import sys
sys.path.append('../utils')
import visualization as visualization
import generic as generic
import project_utils as project_utils
import numpy as np
import matplotlib.pyplot as plt
import importlib

```

### Constants

```python
epidemic_years = ['2012_13', '2015_16', '2018_19', '2023_24']
```

### Load Data

```python
dengue_data = pd.read_csv('../data/processed/dengue_data_with_coordinates.csv', dtype={"semepid":int})
ovitraps_data = pd.read_csv('../data/processed/ovitraps_data_with_coordinates.csv')
```

### Process Data

```python
# Dengue data
biweekly_dengue_data = project_utils.get_biweekly_dengue(dengue_data)
daily_dengue_cases = project_utils.get_daily_dengue(dengue_data)

# ovitraps data
biweekly_ovitraps = project_utils.get_biweekly_ovitraps(ovitraps_data)
biweekly_ovitraps_sum = biweekly_ovitraps.sum(axis=1)
biweekly_ovitraps_mean = biweekly_ovitraps.mean(axis=1)
```

### Plot data

#### Dengue

```python
daily_dengue_cases.plot(kind='line', figsize=(10, 5), title='Dengue daily cases over time')
plt.show()
biweekly_dengue_data.plot(kind='line', figsize=(10, 5), title='Dengue cases by biweek')
plt.show()
```

#### Ovitraps

```python
biweekly_ovitraps_sum.plot(kind='line', figsize=(10, 5), title='Ovitraps count by biweek')
plt.show()
biweekly_ovitraps_mean.dropna().plot(kind='line', figsize=(10, 5), title='Ovitraps average by biweek')
plt.show()
```

# Correlation Analysis

### All years

```python
generic.plot_cross_correlation(
    series_1 = biweekly_ovitraps_mean,
    series_2 = biweekly_dengue_data,
    max_lag =  50,
    title = 'Cross-correlation between ovitraps mean and dengue cases',
    x_label = 'Lags (biweeks)',
    )

generic.plot_cross_correlation(
    series_1 = biweekly_ovitraps_mean,
    series_2 = biweekly_dengue_data,
    max_lag =  20,
    title = 'Cross-correlation between ovitraps mean and dengue cases',
    x_label = 'Lags (biweeks)',

    )
```

### Epidemic years


```python
epidemic_mask = biweekly_dengue_data.index.map(lambda x: x[:7] in epidemic_years)
biweekly_dengue_data_epidemic = biweekly_dengue_data[epidemic_mask]
biweekly_dengue_data_epidemic.plot(kind='line', figsize=(10, 5), title='Dengue cases by biweek (epidemic years)')
```

```python
generic.plot_cross_correlation(
    series_1 = biweekly_ovitraps_mean,
    series_2 = biweekly_dengue_data_epidemic,
    max_lag =  50,
    title = 'Cross-correlation between ovitraps mean and dengue cases (Epidemic Years)',
    x_label = 'Lags (biweeks)',
    )

generic.plot_cross_correlation(
    series_1 = biweekly_ovitraps_mean,
    series_2 = biweekly_dengue_data_epidemic,
    max_lag =  20,
    title = 'Cross-correlation between ovitraps mean and dengue cases (Epidemic Years)',
    x_label = 'Lags (biweeks)',
    )
```

### Non epidemic years

```python
biweekly_dengue_data_non_epidemic = biweekly_dengue_data[~epidemic_mask]
biweekly_dengue_data_non_epidemic.plot(kind='line', figsize=(10, 5), title='Dengue cases by biweek (epidemic years)')
```

```python
# Filter non epidemic years
generic.plot_cross_correlation(
    series_1 = biweekly_ovitraps_mean,
    series_2 = biweekly_dengue_data_non_epidemic,
    max_lag =  50,
    title = 'Cross-correlation between ovitraps mean and dengue cases (non-epidemic Years)',
    x_label = 'Lags (biweeks)',
    )

generic.plot_cross_correlation(
    series_1 = biweekly_ovitraps_mean,
    series_2 = biweekly_dengue_data_non_epidemic,
    max_lag =  20,
    title = 'Cross-correlation between ovitraps mean and dengue cases (non-epidemic Years)',
    x_label = 'Lags (biweeks)',
    )
```

### Correlation by year

```python
all_years = dengue_data.anoepid.unique()
all_years.sort()
max_correlation_dict = {}
for year in all_years:
    year_mask = biweekly_dengue_data.index.map(lambda x: x[:7] == year)
    year_data = biweekly_dengue_data[year_mask]
    try:
        if year in epidemic_years:
            marker = '*'
        else:
            marker = ''    
        # Plot cross-correlation for each year
        generic.plot_cross_correlation(
            series_1 = biweekly_ovitraps_mean,
            series_2 = year_data,
            max_lag =  20,
            title = f'Cross-correlation between ovitraps mean and dengue cases in {year}{marker}',
            x_label = 'Lags (biweeks)',
            )
        plt.show()
        
        # Get the maximum correlation for each year
        max_correlation_dict[year] = generic.max_correlation(
                                    series_1 = biweekly_ovitraps_mean,
                                    series_2 = year_data,
                                    max_lag =  50,
                                            )
    except ValueError as e:
        print(f"Skipping year {year} due to error: {e}")
        continue
```

## Maximum correlation

```python
# Plot max correlations for all years
plt.figure(figsize=(10, 5))
plt.scatter(max_correlation_dict.keys(), [v[1] for v in max_correlation_dict.values()], color='blue', label='All Years')
plt.ylabel('Max Correlation')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.legend()
plt.title('Max Correlation between Ovitraps Mean and Dengue Cases by Year')
plt.show()
```

# Box-plot

```python
biweekly_ovitraps_mean_shifted = biweekly_ovitraps_mean.shift(4)
```

## All years

```python
filtered_df_all = pd.DataFrame()
filtered_series = []
for down,up in zip(range(0,100,10), range(10,110,10)):
    if up > 100:
        up = np.inf
    filter_index = biweekly_ovitraps_mean_shifted[(biweekly_ovitraps_mean_shifted < up) & (biweekly_ovitraps_mean_shifted >= down)].dropna().index
    filtered_series.append(pd.Series(biweekly_dengue_data.loc[biweekly_dengue_data.index.intersection(filter_index)].dropna(),name=f'{down}-{up}'))

filtered_df_all = pd.DataFrame(filtered_series).T
cat_samples = filtered_df_all.notna().sum()

# Plot
plt.figure(figsize=(12, 6))
ax = filtered_df_all.boxplot()
# Add text above each box
for i, col in enumerate(filtered_df_all.columns, start=1):
    ax.text(
        i,                      # x = box position
        filtered_df_all[col].max(),  # y = max value of that box
        str(cat_samples[col]),       # text = count
        ha='center', va='bottom',
        fontsize=10, fontweight='bold', color='darkred'
    )
plt.title('Boxplot of Dengue Cases filtered by Ovitraps - All years')
plt.show()

plt.figure(figsize=(12, 6))
ax = filtered_df_all.clip(upper=1000).boxplot()
# Add text above each box
for i, col in enumerate(filtered_df_all.columns, start=1):
    ax.text(
        i,                      # x = box position
        filtered_df_all[col].clip(upper=1000).max(),  # y = max value of that box
        str(cat_samples[col]),       # text = count
        ha='center', va='bottom',
        fontsize=10, fontweight='bold', color='darkred'
    )
plt.title('Boxplot of Dengue Cases filtered by Ovitraps (truncated at 1000) - All years')
plt.show()
```
