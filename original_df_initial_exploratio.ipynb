## Processing

### Convert type of columns to be unique

```python
df_banco['VALOR_COORDENADA_X'] = df_banco['VALOR_COORDENADA_X'].apply(str)
df_banco['VALOR_COORDENADA_Y'] = df_banco['VALOR_COORDENADA_Y'].apply(str)
```

### Remove useless columns 

```python
df_banco.drop(['Unknown','DT_DESATIVACAO_ARMADILHA_ENDERECO','DES_MUNICIPIO','UF','NUM_SEQU_LOGRADOURO'], axis=1, inplace=True)
# IND_OFICIAL char(1),[Indicativo se o endereço é oficial ou informado.]
```

### NaN Treatment

```python
df_banco.isna().sum().sort_values(ascending=False)
```

NaN means value 0

```python
# IND_QTE_OVOS_ECLODIDOS, IND_QTE_OVOS_DESIDRATADOS
print('Possible values of eclod and desid:\n',
      'IND_QTE_OVOS_ECLODIDOS', df_banco['IND_QTE_OVOS_ECLODIDOS'].unique(),'\n',
      'IND_QTE_OVOS_DESIDRATADOS:', df_banco['IND_QTE_OVOS_DESIDRATADOS'].unique())

#Substituting NaN values with 0
df_banco[['IND_QTE_OVOS_ECLODIDOS','IND_QTE_OVOS_DESIDRATADOS']] = df_banco[['IND_QTE_OVOS_ECLODIDOS','IND_QTE_OVOS_DESIDRATADOS']].fillna(0)


```

### NaN means lack of crutial information

```python
# DT_COLETA
df_banco[df_banco['DT_COLETA'].isna()]
```

```python
# Remove columns with NaN values
df_banco.dropna(subset=['DT_COLETA'], inplace=True)
print('Number of samples after removing NaN values: ', original_shape[0] - df_banco.shape[0])
```
